# Ethical Graphing Notebook

This notebook creates a few basic graphs with Python. The sample data is anonymized and aggregate-only, so it can support a social-study discussion without naming sellers, linking sources, or helping anyone buy fake engagement.

## 1. Import Required Libraries

Import matplotlib and supporting libraries needed for creating graphs.

```python
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["figure.figsize"] = (9, 5)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 11
```

## 2. Create Sample Data

Create small sample datasets that can be reused across different graph examples. Replace these values with your own aggregated, non-identifying observations when you have them.

```python
years = [2019, 2020, 2021, 2022, 2023, 2024]
median_prices = [18, 21, 24, 29, 35, 42]
listing_counts = [8, 11, 15, 21, 28, 36]

source_categories = [
    "Freelance\nmarketplaces",
    "Social-growth\nshops",
    "Marketing\nagencies",
    "Forum\nlistings",
    "Automation\nclaims",
    "Unclear\norganic claims",
]
category_counts = [18, 14, 7, 11, 9, 13]

project_activity = [12, 18, 25, 31, 37, 44, 52, 61]
advertised_price = [16, 19, 23, 24, 31, 34, 41, 48]
```

## 3. Plot a Line Graph

Use matplotlib to create a simple line graph showing values changing over time.

```python
plt.figure()
plt.plot(years, median_prices, marker="o", color="#2563eb", linewidth=2.5)
plt.title("Median Advertised Price Trend")
plt.xlabel("Year")
plt.ylabel("Median advertised price per 100 stars (sample USD)")
plt.xticks(years)
plt.grid(True, alpha=0.35)
plt.show()
```

## 4. Plot a Bar Graph

Create a bar graph to compare values across categories.

```python
plt.figure(figsize=(10, 5))
plt.bar(source_categories, category_counts, color="#16a34a")
plt.title("Observed Source Category Popularity")
plt.xlabel("Source category")
plt.ylabel("Observed count in sample dataset")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

## 5. Plot a Scatter Plot

Create a scatter plot to show the relationship between two numeric variables.

```python
plt.figure()
plt.scatter(project_activity, advertised_price, color="#dc2626", s=80, alpha=0.8)
plt.title("Project Activity vs Advertised Price")
plt.xlabel("Observed project activity score (sample)")
plt.ylabel("Advertised price per 100 stars (sample USD)")
plt.grid(True, alpha=0.35)
plt.show()
```

## 6. Customize Graph Styles

Add titles, axis labels, legends, colors, markers, and grid lines to improve readability.

```python
plt.figure(figsize=(10, 5))
plt.plot(years, median_prices, marker="D", color="#7c3aed", linewidth=2.5, label="Median price")
plt.plot(years, listing_counts, marker="s", color="#ea580c", linewidth=2.5, label="Observed listings")
plt.title("Styled Comparison of Price and Listing Count")
plt.xlabel("Year")
plt.ylabel("Sample value")
plt.xticks(years)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

## 7. Create Multiple Graphs in One Figure

Use subplots to display multiple graph types in a single figure.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].plot(years, median_prices, marker="o", color="#2563eb", linewidth=2)
axes[0].set_title("Line Graph")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Price")
axes[0].grid(True, alpha=0.3)

axes[1].bar(source_categories, category_counts, color="#16a34a")
axes[1].set_title("Bar Graph")
axes[1].set_xlabel("Source category")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis="x", labelrotation=35)
axes[1].grid(axis="y", alpha=0.3)

axes[2].scatter(project_activity, advertised_price, color="#dc2626", s=70, alpha=0.8)
axes[2].set_title("Scatter Plot")
axes[2].set_xlabel("Activity score")
axes[2].set_ylabel("Price")
axes[2].grid(True, alpha=0.3)

fig.suptitle("Basic Graph Examples with Shared Sample Data", fontsize=16)
fig.tight_layout()
plt.show()
```
